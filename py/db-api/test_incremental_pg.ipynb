```python
import psycopg2
from faker import Faker
from dataclasses import dataclass, field

import random
import json
import time
```

```python
faker = Faker()
DBNAME = "forum"
```

```python
@dataclass
class Purchase:
    user_name: str = field(default_factory=faker.user_name)
    currency: str = field(default_factory=faker.currency_code)
    amount: int = field(default_factory=lambda: random.randint(10000, 20000))

    def serialize(self):
        return json.dumps(
            {
                "user_name": self.user_name,
                "currency": self.currency,
                "amount": self.amount,
            }
        )
```

```python
def generate_data():
    
    d = Purchase().serialize()           

    field1 = json.loads(d)['user_name']    
    field2 = json.loads(d)['currency']
    field3 = json.loads(d)['amount']

    return field1, field2, field3
```

```python
generate_data()
# Purchase().serialize()
```

```python
def insert_data(values, type):
    db = psycopg2.connect(dbname=DBNAME, user="postgres", password="postgres")
    c = db.cursor()
    if type == 'serial':
        query = f"insert into purchase_serial(user_name, currency, amount)  values {values}".replace("((","(").replace("))",")")
        c.execute(query)
    elif type == 'sequence':
        seq = " nextval('seq_purchase_id'), "
        # query = "insert into purchase_seq values %s", (values,)
        query = f"insert into purchase_seq values {values}".replace("((","(").replace("))",")")
        query = query.replace("(","({}".format(seq))        
        c.execute(query)
    else:
        query = f"insert into purchase_identify(user_name, currency, amount)  values {values}".replace("((","(").replace("))",")")
        c.execute(query)
    db.commit()
    db.close()

```

```python
def main(type):
    print('Processando...')
    try:
        values = tuple(generate_data() for v in range(10000))
        insert_data(values, type)            
        time.sleep(0.001)
    except KeyboardInterrupt as e:
        print("shutting down")
           
    print('Acabou!')
```

```python
if __name__ == "__main__":
    for _ in range(100):
        main('serial') 
```
